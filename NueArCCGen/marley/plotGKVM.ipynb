# Plot the GKVM Supernova Neutrino Flux

The ROOT file is copied from the DUNE official one, which comes from SNOwGLoBES

```python
import ROOT
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = '/Users/yuntse/data/supernova/gkvm/gkvm.root'
f = ROOT.TFile( filename, 'READ')
```

```python
gGKVMnue = f.Get("nue_E")
```

```python
canvas = ROOT.TCanvas("c1", "GKVM nue E")
gGKVMnue.Draw()
canvas.Draw()
```

```python
gGKVMnue.GetN()
```

## Fill the TGraph into a numpy array

```python
e = np.array(gGKVMnue.GetX())
e
```

```python
len(e)
```

```python
e-0.1
```

```python
lumi = np.array(gGKVMnue.GetY())
np.set_printoptions(formatter={'float': lambda x: f"{x}"})
lumi
```

```python
len(lumi)
```

## Rebinning - Not used in this time

```python
lumi100 = lumi.reshape(500//5, 5).sum(axis =1)
lumi100
```

```python
e100 = np.arange(0.75, 100.75, 1.)
e100
```

```python
fig, ax = plt.subplots( figsize = ( 4, 3), layout = 'tight')

ax.plot( e, lumi, lw = 2, label = 'original' )
ax.plot( e100, lumi100/5, lw = 2, label = 'rebinned' )
ax.set_xlabel( r'E$_\nu$ [MeV]')
ax.set_ylabel( 'Events')
ax.set_xlim( [0, 101])

ax.grid(':')
ax.legend()
```
